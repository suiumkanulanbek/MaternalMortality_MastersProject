```python
import pandas as pd

df = pd.read_excel("uninsured_women.xlsx")

df.head()
```

```python
avg_uninsured_by_state = df.groupby("state")["uninsured women"].mean().sort_values(ascending=False)
avg_uninsured_by_state
```

```python
import matplotlib.pyplot as plt

avg_uninsured_by_state = (
    df.groupby("state")["uninsured women"]
    .mean()
    .reset_index()
    .sort_values(by="uninsured women", ascending=False)
)

plt.figure(figsize=(10, 6))
plt.bar(avg_uninsured_by_state["state"], avg_uninsured_by_state["uninsured women"])
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Average % Uninsured Women")
plt.title("Average Percentage of Uninsured Women by State")
plt.tight_layout()
plt.show()
```
