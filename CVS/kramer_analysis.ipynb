```python
import pandas as pd
import os
import glob


folder_path = 'KramerProject'

csv_files = glob.glob(os.path.join(folder_path, '*.csv'))

df_list = [pd.read_csv(file) for file in csv_files]
merged_df = pd.concat(df_list, ignore_index=True)

merged_df.head(15)
```

```python
merged_df.info()
```

```python
merged_df.to_csv('merged_kramer_data.csv', index=False)
```

```python
merged_df['state'].nunique()
```

```python
#Primary care physician's per capita [county] —  PCPrate 

state_pcprate_avg = merged_df.groupby('state')['PCPrate'].mean().reset_index()
state_pcprate_avg = state_pcprate_avg.sort_values(by='PCPrate', ascending=False)
state_pcprate_avg


#❓ to Dhrumil -- can I count the state average like this? -- just counting average counties's rate?
```
